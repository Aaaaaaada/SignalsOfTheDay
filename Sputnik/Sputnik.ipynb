<h1>The Signal of the Day: The First Man-made Signal Coming in from Outer Space</h1>


<a href="http://en.wikipedia.org/wiki/Sputnik_1">Sputnik-1</a> was the first artificial satellite, launched by the Soviet Union on Oct. 4 1957.

It emitted a regular beep-beep sound downlodable from <a href="http://www.amsat.org/amsat/features/sounds/sputnk1b.wav">here</a>.

By the time the sound reached the earth and due to the performance of the recording equipment at the time 
(analog tape recorders), the sound is noisy, but the regular beep-beep can be clearly heard.
There was no information carried by this signal, just its existence was fascinating. 

<img src="http://upload.wikimedia.org/wikipedia/commons/b/be/Sputnik_asm.jpg" width="600"></img>


<h2>Significance of the Signal</h2>

The fact that satellites can be place into orbit (in this case, it flew for about 3 months before crashing) and can send signals back to earth showed the possibility of telecommunication satellites  to do radio/television transmission, and can carry telephone conversations across oceans. This was revolutionary at the time, and it would take another decade to become reality, while it is commonplace today.

Of course, launching a satellite also showed the possibility of space exploration, starting in effect a race between the Soviet Union and the USA to place cosmonauts (Soviet version) or astronauts (US version) into orbit. This was achieved by <a href="http://en.wikipedia.org/wiki/Yuri_Gagarin">Gagarin</a> (a cosmonaut) in April 1961, and then <a href="http://en.wikipedia.org/wiki/Alan_Shepard">Shepard</a> (an astronaut) in May 1961. The race to the moon that followed was won by the Apollo 11 team, landing on Earth's satellite in July 1969.

<img src="http://www.ast.cam.ac.uk/sites/default/files/assets/images/outreach/gagarin.jpg" width=300></a>

Another, cultural effect was the <a href="http://en.wikipedia.org/wiki/Sputnik_crisis"><i>Sputnik crisis</i></a> where the american government became aware of the urgency of funding for science education so as to not lag behind the Soviet Union in math and science, which was critical in the cold war effort. This lead to increased funding, and improved education and research in math, science and engineering.


<h2>Playing with the Signal of the Day</h2>

Load the signal and print its sampling rate

```python
from scipy.io import wavfile
fs, data = wavfile.read("sputnk1b.wav")
print fs
```

Plot the waveform

```python
%pylab inline
plot(data)
xlabel("Time")
ylabel("Amplitude")
```

Plot the waveform in the spectral domain. Determine the frequency of the sinusoid.

```python
%pylab inline
data_frequency_domain=np.fft.fftpack.fft(data);
plot(abs(data_frequency_domain))
xlabel("Frequency")
ylabel("Amplitude")
```

Plot the short-time fourier transform (STFT) of the signal.

(We use the STFT code proposed <a href="http://stackoverflow.com/questions/2459295/stft-and-istft-in-python">here</a>)

```python
%pylab inline
import scipy
def stft(x, fs, framesz, hop):
    framesamp = int(framesz*fs)
    hopsamp = int(hop*fs)
    w = scipy.hamming(framesamp)
    X = scipy.array([scipy.fft(w*x[i:i+framesamp]) 
                     for i in range(0, len(x)-framesamp, hopsamp)])
    return X

frame_size = 0.05
time_shift = 0.020 

X = stft(data, fs, frame_size, time_shift)

pylab.figure()
pylab.imshow(abs(X.T), origin='lower', aspect='auto',interpolation='nearest')
pylab.xlabel('Time')
pylab.ylabel('Frequency')
pylab.show()
```

You can change the parameters <code>frame_size</code> (the length of every frame to which a traditional Fourier transform is applied) and <code>time_shift</code> (the distance between every two consecutive frames) to see the effects.

Try to denoise the signal.

```python
y=np.zeros((len(data),1))
k=10;
for i in range(0,k+1):
    y[i]=data[i];
for i in range(k+1,len(data)): # (H-red)
    _sum=0.0;   
    for j in range(0,k+1):  
        _sum=_sum+data[i-j];
    y[i]=_sum/k;
subplot(3,3,1)   
plot(data)
xlabel("Time")
ylabel("Amplitude")
subplot(3,3,3)
plot(abs(scipy.fft(data)))
xlabel("Frequency")
ylabel("Amplitude")

subplot(3,3,7)   
plot(y)
xlabel("Time")
ylabel("Amplitude")

subplot(3,3,9)
plot(abs(scipy.fft(y)))
xlabel("Frequency")
ylabel("Amplitude")

scaled = np.int16(y/np.max(np.abs(y)) * 32767)

wavfile.write("denoised.wav",fs,scaled)
```

We denoise the signal using a simple averaging filter. You can view the effect of the order of the signal (<code>k</code>) by changing its value.
