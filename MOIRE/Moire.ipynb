Signal of the Day: The Moire Pattern
====================================

Introduction
------------

It may have happened for you several times, especially if you are a male: You are wearing a casual striped shirt and you decide to appear in a photo. After you look at the taken photo, you notice some weird patterns on your shirt in the photo. Look at the following image for example:

![](http://www.btobey.com/nikon/images/colormoire.jpg)

Of course there is nothing unusual with this shirt or yours. What has happened is an example of a phenomenon called the _Moirè pattern_.

As a signal processing student, you can think of the Moire pattern as the realizations of aliasing in the space of 2-
D image signals. Here, the synthetic frequencies which are created due to the inappropriate subsampling, appear as periodic pattern in the down-sampled image. That's the reason that the frequency of these synthetic pattern is lower than the minimum frequency present in the image itself.

Moire by Downsampling
---------------------

Moire patterns are usually best visible in images with highly repeated patterns. Consider the following photo for example:

```python
import Image, ImageDraw, ImageStat
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
from PIL.ImageDraw import Draw

from io import BytesIO

from IPython.core import display
from PIL import Image


def display_pil_image(im):
   """Displayhook function for PIL Images, rendered as PNG."""

   b = BytesIO()
   im.save(b, format='png')
   data = b.getvalue()

   ip_img = display.Image(data=data, format='png', embed=True)
   return ip_img._repr_png_()


# register display func with PNG formatter:
png_formatter = get_ipython().display_formatter.formatters['image/png']
dpi = png_formatter.for_type(Image.Image, display_pil_image)

im = Image.open("brick.png")

im
```

The bricks are readily visible here. Now we reduce the size of the image by a factor of 2 in each dimension:

Still we have a decent photo. Downsampling by another factor of 2 however,

The fact that Moire patterns appear if we down-sample the signal more than a pre-defined threshold, is a strong evidence to verify our hypothesis that this is related to sampling and the frequency domain. To make sure, let's see what happens in the Fourier domain when we down-sample the brick image.

The following code computes and shows the Fourier domain for various down-sampled versions of the brick image and the original image as well.

```python
%pylab inline

def fftshow(im):
    im5a=np.asarray(im);

    im5f=np.fft.fft2(im5a[:,:,0]);

    im5f[0,0]=0;

    plt.imshow(np.log(np.abs(np.fft.fftshift(im5f))),aspect='auto',interpolation='none')


fftshow(im)
```

```python
fftshow(im2)
```

```python
fftshow(im4)
```

```python
fftshow(im8)
```

Please note that we remove the DC component of the Fourier transform so that the Fourier image becomes more visible.

We can see in the images above the frequency components that lie a little far from the center of the image and hence cause aliasing when downsampling the image. Interference of these frequencies causes the Moire patterns.

Generating Moire Patterns
------------------------

Now let's see more closely how to create a Moire pattern. We first form a simple image containing a periodic set of stripes (similar to a pulse train in 1D) with period $p$:


```python
def bars(L,p,width):
    bar=np.zeros((L,L,3),'uint8')
    for i in np.arange(0,L):
        for j in np.arange(p,L-p,p):
            for k in np.arange(j,j+width):
                bar[i,k]=(255,255,255);
    return Image.fromarray(bar);
    
imma=bars(1024,48,8);

plt.imshow(np.asarray(imma),aspect='auto')
```

We create also a second image, with a period $p \leq q \leq 2p$:

```python
immb=bars(1024,64,8);

plt.imshow(np.asarray(immb),aspect='auto',interpolation='none')
```

Now, combining the two patterns by superimposing them, we see a new pattern (the Moire effect):

```python
immc=Image.blend(imma,immb,0.5);

plt.imshow(np.asarray(immc),aspect='auto',interpolation='none')
```

Try to think why superimposing causes aliasing?

Now, we rotate one of the patterns and superimpose them to see what happens:

```python
immd5=immb.rotate(5);

imme5=Image.blend(imma,immd5,0.5);

plt.imshow(np.asarray(imme5),aspect='auto',interpolation='none')
```

You can see the appeared diamond-shaped patterns. We can easily show that the side  of each diamond equals $\frac{p}{\sin\alpha}$ where $p$ is the smaller period and $\alpha$ is the rotation angle. 

More serious Moire patternsappear at smaller rotation angles. Let's verify this:

```python
immd10=immb.rotate(10);

imme10=Image.blend(imma,immd10,0.5);

plt.imshow(np.asarray(imme10),aspect='auto',interpolation='none')
```

```python
immd25=immb.rotate(25);

imme25=Image.blend(imma,immd25,0.5);

plt.imshow(np.asarray(imme25),aspect='auto',interpolation='none')
```

```python
immd45=immb.rotate(45);

imme45=Image.blend(imma,immd45,0.5);

plt.imshow(np.asarray(imme45),aspect='auto',interpolation='none')
```

Overlaying in the Frequency Domain
----------------------------------

Now let's see what happens in the frequency domain. 

We know from theory that rotation in the image plane causes rotation in the frequency plane. Let's verify this first:

```python
subplot(221)
fftshow(immd5)
title('5 degree rotation')

subplot(222)
fftshow(immd10)
title('10 degree rotation')

subplot(223)
fftshow(immd25)
title('25 degree rotation')

subplot(224)
fftshow(immd45)
title('45 degree rotation')
```

Then, we try to figure out how superimposition affects the frequency plane:

```python
subplot(221)
fftshow(imme5)
title('Overlaid with 5 degree rotation')

subplot(222)
fftshow(imme10)
title('Overlaid with 10 degree rotation')

subplot(223)
fftshow(imme25)
title('Overlaid with 25 degree rotation')

subplot(224)
fftshow(imme45)
title('Overlaid with 45 degree rotation')
```

What happens is that superimposition has an effect similar to multiplication in the spatial domain, which causes convultion in the frequency domain, hence introducing new frequencies.

Moire Patterns in Printing
--------------------------

Moire patterns have applications in the printing industry as well. In many printing systems, all what we can do is monochrome printing, that is printing either a white pixel or a black one, though with a very high resolution. One approach to create the effect of gradients using only two colors, is by using the so-called "Half-tone imaging".

In halftone imaging, screens containing patterns of dots are overlaid on each other in order to reproduce multiple gray levels and colors. Let's try to do so now.

We first create a grid of dot patterns and a rotated copy of it: 

```python
def grid(L,spacing,R):
    grid=np.ones((L,L,3),'uint8')*255
    for i in np.arange(spacing,L-spacing,spacing):
        for j in np.arange(spacing,L-spacing,spacing):
            for k in np.arange(i-R,i+R):
                for l in np.arange(j-R,j+R):
                    if np.sqrt((k-i)**2+(l-j)**2)<=R:
                        grid[k,l]=(255,0,255);
    return Image.fromarray(grid)
        
grid=grid(512,8,3);

    
grid_rotated=im7.rotate(15);
```

We then overlay the two patterns and see the results:

```python

pattern=Image.blend(grid,grid_rotated,0.5);

plt.imshow(np.asarray(pattern),aspect='auto',interpolation='none')
```

The patterns you can see in the above image are called _Rosettes_ and widely known in the printing industry. 

If you look carefully from close distance to images printed on a newspaper for example, you can distinguish Rosette patterns. This is because newspapers are usually printed using half-tone imaging.

The following code creates a half-tone effect on an image: (Code borrowed from _here_)

```python
def gcr(im, percentage):
    '''basic "Gray Component Replacement" function. Returns a CMYK image with 
       percentage gray component removed from the CMY channels and put in the
       K channel, ie. for percentage=100, (41, 100, 255, 0) >> (0, 59, 214, 41)'''
    cmyk_im = im.convert('CMYK')
    if not percentage:
        return cmyk_im
    cmyk_im = cmyk_im.split()
    cmyk = []
    for i in xrange(4):
        cmyk.append(cmyk_im[i].load())
    for x in xrange(im.size[0]):
        for y in xrange(im.size[1]):
            gray = min(cmyk[0][x,y], cmyk[1][x,y], cmyk[2][x,y]) * percentage / 100
            for i in xrange(3):
                cmyk[i][x,y] = cmyk[i][x,y] - gray
            cmyk[3][x,y] = gray
    return Image.merge('CMYK', cmyk_im)

def halftone(im, cmyk, sample, scale):
    '''Returns list of half-tone images for cmyk image. sample (pixels), 
       determines the sample box size from the original image. The maximum 
       output dot diameter is given by sample * scale (which is also the number 
       of possible dot sizes). So sample=1 will presevere the original image 
       resolution, but scale must be >1 to allow variation in dot size.'''
    cmyk = cmyk.split()
    dots = []
    angle = 0
    for channel in cmyk:
        channel = channel.rotate(angle, expand=1)
        size = channel.size[0]*scale, channel.size[1]*scale
        half_tone = Image.new('L', size)
        draw = ImageDraw.Draw(half_tone)
        for x in xrange(0, channel.size[0], sample):
            for y in xrange(0, channel.size[1], sample):
                box = channel.crop((x, y, x + sample, y + sample))
                stat = ImageStat.Stat(box)
                diameter = (stat.mean[0] / 255)**0.5
                edge = 0.5*(1-diameter)
                x_pos, y_pos = (x+edge)*scale, (y+edge)*scale
                box_edge = sample*diameter*scale
                draw.ellipse((x_pos, y_pos, x_pos + box_edge, y_pos + box_edge), fill=255)
        half_tone = half_tone.rotate(-angle, expand=1)
        width_half, height_half = half_tone.size
        xx=(width_half-im.size[0]*scale) / 2
        yy=(height_half-im.size[1]*scale) / 2
        half_tone = half_tone.crop((xx, yy, xx + im.size[0]*scale, yy + im.size[1]*scale))
        dots.append(half_tone)
        angle += 15
    return dots


im = Image.open("lena.png")

cmyk = gcr(im, 0)
dots = halftone(im, cmyk, 5, 10)
newsp = Image.merge('CMYK', dots)
```
