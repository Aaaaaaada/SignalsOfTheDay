<h1>Signal of the Day: Lehman Brothers' Stock Value</h1>

This signal shows the  daily stock values of Lehman Brothers' firm during the year 2008, in which it declared bankruptcy.  The values can be downloaded from <a href="http://charting.nasdaq.com/ext/charts.dll?2-1-14-0-0-512-03NA000000LEHLQ-&SF:1%7C5-WD=530-HT=395--XTBL-">here</a>.

Lehman Brothers Holdings Inc. (former NYSE ticker symbol LEH)  was a global financial services firm. Before declaring bankruptcy in 2008, Lehman was the fourth-largest investment bank in the US (behind Goldman Sachs, Morgan Stanley, and Merrill Lynch), doing business in investment banking, equity and fixed-income sales and trading (especially U.S. Treasury securities), research, investment management, private equity, and private banking.

At 1:45AM on September 15, 2008, the firm filed for Chapter 11 bankruptcy protection following the massive exodus of most of its clients, drastic losses in its stock, and devaluation of its assets by credit rating agencies.

<h1>Importance of the Signal</h1>

With \$639 billion in assets and \$619 billion in debt, Lehman's bankruptcy filing was the largest in history, as its assets far surpassed those of previous bankrupt giants such as WorldCom and Enron. Lehman was the fourth-largest U.S.investment bank at the time of its collapse, with 25,000 employees worldwide. The bankruptcy of the Lehman Brothers is thought to have played a major role in the unfolding of the late-2000s global financial crisis.

Lehman's demise also made it the largest victim of the U.S. sub-prime mortgage-induced financial crisis that swept through global financial markets in 2008. Lehman's collapse was a seminal event that greatly intensified the 2008 crisis and contributed to the erosion of close to $10 trillion in market capitalization from global equity markets in October 2008, the biggest monthly decline on record at the time.

<h2>Playing with the Signal of the Day</h2>

Read the signal from the data file. 

```python
%pylab inline
import datetime as dt

fp = open('LehmanBrothers_data.txt')
columns = ['Close/Last','Volume']
data = empty((0,len(columns)))
dates = []
for line in fp.readlines()[1:]:
    fields = line.split('\t')
    dates.append(dt.datetime.strptime(fields[0], "%Y/%m/%d"))
    data = vstack([data, array([[float(fields[1].replace(',','')), float(fields[2].replace(',',''))]])])
fp.close()
```

Show the first few values in a tabular form.

```python
import pandas as pd
df = pd.DataFrame(data, index=dates, columns=columns)
df.head()
```

Plot the values in a time-series form and notice the abrupt change at the time of bankruptcy.

```python
subplot(311)
df['Close/Last'].plot()
ylabel('Value')
xlabel('Date');
subplot(313)
df['Volume'].plot()
ylabel('Volume')
xlabel('Date');
```
